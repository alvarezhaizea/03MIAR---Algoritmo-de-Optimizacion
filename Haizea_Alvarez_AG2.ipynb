```python
#N Reinas - Vuelta Atrs()
################################################################################

#Verifica que en la solucin parcial no hay amenzas entre reinas
################################################################
def es_prometedora(SOLUCION,etapa):
################################################################
  #print(SOLUCION)
  #Si la solucin tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " est " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True

#Traduce la solucin al tablero
################################################################
def escribe_solucion(S):
################################################################
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")



#Proceso principal de N-Reinas
################################################################
def reinas(N, solucion=[],etapa=0):
################################################################
### ....
  if len(solucion) == 0:         # [0,0,0...]
    solucion = [0 for i in range(N) ]

  for i in range(1, N+1):
    solucion[etapa] = i
    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

  solucion[etapa] = 0

reinas(5,solucion=[],etapa=0)
#N numero de reinas
```

```python
def distancia_minima_fuerza_bruta_1D(lista):
  d_min = float('inf')
  punto1 = None
  punto2 = None

  # Recorremos todos los pares de puntos
  for i in range(0, len(lista)):
    for j in range(i + 1,len(lista)):
      # Calculamos la distancia entre lista[i] y lista[j]
      distancia = abs(lista[i]-lista[j])

      # Actualizamos la distancia mnima si encontramos una menor
      if d_min>distancia:
        d_min = distancia
        punto1 = lista[i]
        punto2 = lista[j]
  return [punto1,punto2], d_min
```

```python
def distancia_minima_divide_venceras_1D(lista):
    # Ordenar la lista de entrada
    lista_ordenada = sorted(lista)
    n = len(lista_ordenada)

    # Si hay menos de 2 elementos, devolvemos una tupla vlida aunque no haya suficientes puntos
    if n < 2:
        return [None, None], float('inf')

    elif n == 2:
        # Caso base: solo dos puntos
        return [lista_ordenada[0], lista_ordenada[1]], abs(lista_ordenada[0] - lista_ordenada[1])

    else:
        # Dividir en dos mitades
        mid = n // 2
        izquierda = lista_ordenada[:mid]
        derecha = lista_ordenada[mid:]

        # Resolver recursivamente para ambas mitades
        pt_min_izq, d_min_izq = distancia_minima_divide_venceras_1D(izquierda)
        pt_min_der, d_min_der = distancia_minima_divide_venceras_1D(derecha)

        # Calcular la distancia mnima en la franja central (si es posible)
        if mid > 0:  # Validar que mid-1 no cause un error de ndice
            d_min_centro = abs(lista_ordenada[mid] - lista_ordenada[mid - 1])
            punto1_centro, punto2_centro = lista_ordenada[mid - 1], lista_ordenada[mid]
        else:
            d_min_centro = float('inf')
            punto1_centro, punto2_centro = None, None

        # Comparar las distancias mnimas y devolver el mejor resultado
        if d_min_izq < d_min_centro and d_min_izq < d_min_der:
            return pt_min_izq, d_min_izq
        elif d_min_centro < d_min_izq and d_min_centro < d_min_der:
            return [punto1_centro, punto2_centro], d_min_centro
        else:
            return pt_min_der, d_min_der
```

```python
def distancia_minima_fuerza_bruta_2D(lista):
    d_min = float('inf')
    punto1 = None
    punto2 = None

    # Recorremos todos los pares de puntos
    for i in range(len(lista)):
        for j in range(i + 1, len(lista)):
            # Calculamos la distancia entre lista[i] y lista[j]
            distancia = math.sqrt((lista[i][0] - lista[j][0])**2 + (lista[i][1] - lista[j][1])**2)

            # Actualizamos la distancia mnima si encontramos una menor
            if distancia < d_min:
                d_min = distancia
                punto1 = lista[i]
                punto2 = lista[j]

    return [punto1, punto2], d_min
```

```python
def distancia_minima_divide_venceras_2D(lista):
    # Ordenar la lista de entrada
    lista_ordenada = sorted(lista)
    n = len(lista_ordenada)

    # Si hay menos de 2 elementos, devolvemos una tupla vlida aunque no haya suficientes puntos
    if n < 2:
        return [None, None], float('inf')

    elif n == 2:
        # Caso base: solo dos puntos
        return [lista_ordenada[0], lista_ordenada[1]], math.sqrt((lista[0][0] - lista[1][0])**2 + (lista[0][1] - lista[1][1])**2)

    else:
        # Dividir en dos mitades
        mid = n // 2
        izquierda = lista_ordenada[:mid]
        derecha = lista_ordenada[mid:]

        # Resolver recursivamente para ambas mitades
        pt_min_izq, d_min_izq = distancia_minima_divide_venceras_2D(izquierda)
        pt_min_der, d_min_der = distancia_minima_divide_venceras_2D(derecha)

        # Calcular la distancia mnima en la franja central (si es posible)
        if mid > 0:  # Validar que mid-1 no cause un error de ndice
            d_min_centro = math.sqrt((lista[mid -1][0] - lista[mid][0])**2 + (lista[mid -1][1] - lista[mid][1])**2)
            punto1_centro, punto2_centro = lista_ordenada[mid - 1], lista_ordenada[mid]
        else:
            d_min_centro = float('inf')
            punto1_centro, punto2_centro = None, None

        # Comparar las distancias mnimas y devolver el mejor resultado
        if d_min_izq < d_min_centro and d_min_izq < d_min_der:
            return pt_min_izq, d_min_izq
        elif d_min_centro < d_min_izq and d_min_centro < d_min_der:
            return [punto1_centro, punto2_centro], d_min_centro
        else:
            return pt_min_der, d_min_der
```

```python
def distancia_minima_fuerza_bruta_3D(lista):
    d_min = float('inf')
    punto1 = None
    punto2 = None

    # Recorremos todos los pares de puntos
    for i in range(len(lista)):
        for j in range(i + 1, len(lista)):
            # Calculamos la distancia entre lista[i] y lista[j]
            distancia = math.sqrt(
                (lista[i][0] - lista[j][0])**2 +
                (lista[i][1] - lista[j][1])**2 +
                (lista[i][2] - lista[j][2])**2
            )

            # Actualizamos la distancia mnima si encontramos una menor
            if distancia < d_min:
                d_min = distancia
                punto1 = lista[i]
                punto2 = lista[j]

    return [punto1, punto2], d_min
```

```python
def distancia_minima_divide_venceras_3D(lista):
    # Ordenar la lista de entrada
    lista_ordenada = sorted(lista)
    n = len(lista_ordenada)

    # Si hay menos de 2 elementos, devolvemos una tupla vlida aunque no haya suficientes puntos
    if n < 2:
        return [None, None], float('inf')

    elif n == 2:
        # Caso base: solo dos puntos
       return [lista_ordenada[0], lista_ordenada[1]], math.sqrt(
            (lista_ordenada[0][0] - lista_ordenada[1][0])**2 +
            (lista_ordenada[0][1] - lista_ordenada[1][1])**2 +
            (lista_ordenada[0][2] - lista_ordenada[1][2])**2
        )

    else:
        # Dividir en dos mitades
        mid = n // 2
        izquierda = lista_ordenada[:mid]
        derecha = lista_ordenada[mid:]

        # Resolver recursivamente para ambas mitades
        pt_min_izq, d_min_izq = distancia_minima_divide_venceras_2D(izquierda)
        pt_min_der, d_min_der = distancia_minima_divide_venceras_2D(derecha)

        # Calcular la distancia mnima en la franja central (si es posible)
        if mid > 0:  # Validar que mid-1 no cause un error de ndice
            d_min_centro = math.sqrt(
                (lista_ordenada[mid - 1][0] - lista_ordenada[mid][0])**2 +
                (lista_ordenada[mid - 1][1] - lista_ordenada[mid][1])**2 +
                (lista_ordenada[mid - 1][2] - lista_ordenada[mid][2])**2
            )
            punto1_centro, punto2_centro = lista_ordenada[mid - 1], lista_ordenada[mid]
        else:
            d_min_centro = float('inf')
            punto1_centro, punto2_centro = None, None

        # Comparar las distancias mnimas y devolver el mejor resultado
        if d_min_izq < d_min_centro and d_min_izq < d_min_der:
            return pt_min_izq, d_min_izq
        elif d_min_centro < d_min_izq and d_min_centro < d_min_der:
            return [punto1_centro, punto2_centro], d_min_centro
        else:
            return pt_min_der, d_min_der
```
